```python
import os, sys
sys.path.append(os.path.abspath(os.path.join('../src/')))
from f_gan import *

# Enable Jupyter notebook plotting
%matplotlib inline
```

```python
# Load in binzarized MNIST data, separate into data loaders
train_iter, val_iter, test_iter = get_data()

# Init model
model = fGAN(image_size=784,
                hidden_dim=400,
                z_dim=20,
                output_dim=1)
```

```python
# Init trainer
trainer = fGANTrainer(model=model,
                    train_iter=train_iter,
                    val_iter=val_iter,
                    test_iter=test_iter,
                    viz=True)

# Train
trainer.train(num_epochs=25,
            method='total_variation',
            G_lr=1e-4,
            D_lr=1e-4,
            D_steps=1)
```

```python
# Show loss
trainer.viz_loss()
```
