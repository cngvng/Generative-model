```python
import os, sys
sys.path.append(os.path.join('../src/'))
from vae import *

# Enable juoyter notebook plotting
%matplotlib inline
```

```python
# Load in binzarized MNIST data, separate into data loaders
train_iter, val_iter, test_iter = get_data()

# Init model
model = VAE(image_size=784,
            hidden_dim=400,
            z_dim=20)
```

```python
# Init trainer
trainer = VAETrainer(model=model,
                    train_iter=train_iter,
                    val_iter=val_iter,
                    test_iter=test_iter,
                    viz=True)

# Train
trainer.train(num_epochs=25,
            lr=1e-3,
            weight_decay=1e-5)
```

```python
# Plot loss
trainer.viz_loss()

# Explore latent space
trainer.make_all
```
