```python
import os, sys
sys.path.append(os.path.abspath(os.path.join('../src/')))
from w_gp_gan import *

# Enable Jupyter notebook plotting
%matplotlib inline
```

```python
# Load in binzarized MNIST data, separate into data loaders
train_iter, val_iter, test_iter = get_data()

# Init model
model = WGPGAN(image_size=784,
             hidden_dim=400,
             z_dim=20)
```

```python
# Init trainer
trainer = WGPGANTrainer(model=model,
                        train_iter=train_iter,
                        val_iter=val_iter,
                        test_iter=test_iter,
                        viz=True)

# Train
trainer.train(num_epochs=25,
            G_lr=5e-5,
            D_lr=5e-5,
            D_steps=5)
```

```python
# Show loss
trainer.viz_loss()
```
